```python
import numpy as np
import cupy as cp#所有数据都在GPU上创建和计算
import os
import pickle
from tqdm import tqdm
from sklearn.model_selection import ParameterSampler
%run /data/nlp-zbj-22/cv-fdu/assignment1/model.ipynb
%run /data/nlp-zbj-22/cv-fdu/assignment1/loss.ipynb
%run /data/nlp-zbj-22/cv-fdu/assignment1/trainer.ipynb
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt

def load_cifar10_with_augmentation(path="/data/nlp-zbj-22/cv-fdu/assignment1/cifar-10-batches-py", seed=42):
    # 加载原始数据
    X_train_full, y_train_full = [], []
    for i in range(1, 6):
        with open(os.path.join(path, f'data_batch_{i}'), 'rb') as f:
            batch = pickle.load(f, encoding='latin1')
            X_train_full.append(cp.array(batch['data'].astype(cp.float32)))
            y_train_full.extend(batch['labels'])
    
    X_train_full = cp.concatenate(X_train_full, axis=0)
    y_train_full = cp.array(y_train_full)

    # 加载测试集
    with open(os.path.join(path, 'test_batch'), 'rb') as f:
        batch = pickle.load(f, encoding='latin1')
        X_test = cp.array(batch['data']).astype(cp.float32)
        y_test = cp.array(batch['labels'])

    # 分层划分验证集（5k），转换为numpy数组进行split(因为cupy数组不能使用sklearn的split）
    X_train_full_np = cp.asnumpy(X_train_full)
    y_train_full_np = cp.asnumpy(y_train_full)
    X_train_np, X_val_np, y_train_np, y_val_np = train_test_split(
        X_train_full_np, y_train_full_np, test_size=5000,
        stratify=y_train_full_np, random_state=seed
    )
    # 转回cupy数组
    X_train = cp.array(X_train_np)
    X_val = cp.array(X_val_np)
    y_train = cp.array(y_train_np)
    y_val = cp.array(y_val_np)
    

    # 通道级标准化（基于训练集统计）
    def normalize(X, mean=None, std=None):
        X = X.reshape(-1, 3, 32, 32).transpose(0,2,3,1)  # [N,32,32,3]
        if mean is None:
            mean = cp.mean(X, axis=(0,1,2))
            std = cp.std(X, axis=(0,1,2))
        X_norm = (X - mean) / (std + 1e-8)
        return X_norm.reshape(-1, 3072), mean, std
    
    X_train_norm, mean, std = normalize(X_train)
    X_val_norm, _, _ = normalize(X_val, mean, std)
    X_test_norm, _, _ = normalize(X_test, mean, std)

    # 离线增强模式（直接生成增强数据）
    # 应用基础增强：水平翻转+随机裁剪
    augmented_X, augmented_y = [], []
    for img, label in zip(X_train_norm, y_train):
        img = img.reshape(32,32,3)
        # 水平翻转
        flipped = img[:, ::-1, :]
        augmented_X.append(flipped.reshape(-1))
        augmented_y.append(label)
        # 随机裁剪（填充后裁剪）
        pad_size = 4
        padded = cp.pad(img, [(pad_size,pad_size)]*2 + [(0,0)], mode='reflect')
        y = int(cp.random.randint(0, 2*pad_size).item())  # 生成Python整数
        x = int(cp.random.randint(0, 2*pad_size).item())
        cropped = padded[y:y+32, x:x+32, :]
        augmented_X.append(cropped.reshape(-1))
        augmented_y.append(label)
        
    # 合并增强数据
    X_train_aug = cp.concatenate([X_train_norm, cp.array(augmented_X)], axis=0)  
    y_train_aug = cp.concatenate([y_train, cp.array(augmented_y)], axis=0)       
    
    # 打乱增强数据
    idx = cp.random.permutation(len(X_train_aug))  # np改为cp
    return X_train_aug[idx], y_train_aug[idx], X_val_norm, y_val, X_test_norm, y_test

X_train, y_train, X_val, y_val, X_test, y_test = load_cifar10_with_augmentation()

# print(f"训练集形状: {X_train.shape}")  # (45000, 3072)
# print(f"验证集形状: {X_val.shape}")    # (5000, 3072)
# print(f"测试集形状: {X_test.shape}")   # (10000, 3072)
# print(y_train.shape)
# print(y_train[0:10])

def evaluate(model, X, y):
    logits = model.forward(X)
    preds = cp.argmax(logits, axis=1)  # 使用cupy的argmax
    acc = cp.mean(preds == y)
    return float(acc)  # 转换为Python浮点数



model = myNet(
    input_dim=3072,
    hidden_dim1=512,
    hidden_dim2=128,
    output_dim=10,
    activation='swish'
)


trainer = SGDTrainer(
    model=model,
    lr=0.001,
    lr_decay=0.98,
    reg_lambda=0.01
)




#训练参数
num_epochs = 200
batch_size = 256

best_acc=float("-inf")
best_val_acc=float("-inf")


# 初始化存储变量（阶段1专用）,用于最后的作图
stage1_train_losses = []
stage1_val_losses = []
stage1_val_accs = []


# 训练循环
for epoch in range(num_epochs):
    epoch_loss = 0.0
    with tqdm(total=len(X_train), desc=f"Epoch {epoch+1}/{num_epochs}", unit="samples") as pbar:
        # Mini-batch训练
        for i in range(0, len(X_train), batch_size):
            # 获取当前batch
            X_batch = X_train[i:i+batch_size]
            y_batch = y_train[i:i+batch_size]
            model.zero_grad()#最开始的时候清理原来剩余的梯度
            logits = model.forward(X_batch)
            loss, grad = cross_entropy_loss(logits, y_batch, model, trainer.reg_lambda)
            epoch_loss += loss * len(X_batch)
            model.backward(grad)
            trainer.step()
            pbar.update(len(X_batch))
    trainer.decay_learning_rate()
    avg_loss = epoch_loss / len(X_train)#训练集的loss
    val_logits = model.forward(X_val)
    val_loss, _ = cross_entropy_loss(val_logits, y_val, model, trainer.reg_lambda)#验证集的loss
    val_acc = evaluate(model, X_val, y_val)#验证集的acc
    print(f"Epoch {epoch+1}/{num_epochs} | Loss: {avg_loss:.4f} | Val Acc: {val_acc:.2%} | Val Loss: {val_loss:.4f}")
    #每一个epoch结束我们就处理一下，看是否有更好的参数要保存
    if val_acc > best_val_acc:
        best_val_acc = val_acc
        model.save_model(stage=1)
        print(f"New best model saved! Acc: {val_acc:.2%}")
        
    # 记录数据（转换为Python浮点数），用于绘图
    stage1_train_losses.append(float(avg_loss))
    stage1_val_losses.append(float(val_loss))
    stage1_val_accs.append(float(val_acc))

# 阶段1可视化（所有epoch跑完之后再绘图）
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.plot(stage1_train_losses, label='Train Loss')
plt.plot(stage1_val_losses, label='Val Loss')
plt.title("Stage1 Loss Curves")
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(stage1_val_accs, label='Val Acc')
plt.title("Stage1 Validation Accuracy")
plt.legend()
plt.show()
    





# 最终测试评估    
model.load_model(stage=1)#去使用我们最好的模型  

# 对模型参数的可视化
linear_idx = 0
for layer in model.layers:
    if isinstance(layer, Linear):
        W = layer.W.get().flatten()  # CuPy数组转NumPy: .get()
        b = layer.b.get().flatten()  # CuPy数组转NumPy: .get()
        
        # 创建新图表
        plt.figure(figsize=(12, 5))
        
        # 绘制权重W的直方图
        plt.subplot(1, 2, 1)
        plt.hist(W, bins=50, color='blue', alpha=0.7)  # 直接使用转换后的NumPy数组
        plt.title(f'Linear {linear_idx} Weights Distribution')
        plt.xlabel('Weight Value')
        plt.ylabel('Frequency')
        
        # 绘制偏置b的直方图
        plt.subplot(1, 2, 2)
        plt.hist(b, bins=50, color='green', alpha=0.7)  # 直接使用转换后的NumPy数组
        plt.title(f'Linear {linear_idx} Biases Distribution')
        plt.xlabel('Bias Value')
        plt.ylabel('Frequency')

        plt.show()
        linear_idx += 1



test_acc = evaluate(model, X_test, y_test)
print(f"\nFinal Test Accuracy: {test_acc:.2%}")


# ##############超参数搜索部分#######################

# 定义超参数范围
param_dist = {
    'lr': [0.01, 0.001, 0.0001],
    'lr_decay': [0.9, 0.95, 0.99],
    'reg_lambda': [0.001, 0.01, 0.1],
    'hidden_dim1': [512, 1024, 2048],
    'hidden_dim2': [256, 512, 1024],
    'batch_size': [64, 128, 256],
    'num_epochs': [100, 200, 300]
}

# 随机采样超参数组合
n_iter = 100 # 采样次数
param_list = list(ParameterSampler(param_dist, n_iter=n_iter))

best_val_acc = 0
best_model_state=None
best_params = None

# 遍历参数组合
for params in tqdm(param_list, desc="Hyperparameter Search"):  # 添加 tqdm
    print(f"\nTraining with params: {params}")
    
    # 初始化模型
    model = myNet(
        input_dim=3072,
        hidden_dim1=params['hidden_dim1'],
        hidden_dim2=params['hidden_dim2'],
        output_dim=10,
        activation='relu'
    )
    
    # 初始化训练器
    trainer = SGDTrainer(
        model=model,
        lr=params['lr'],
        lr_decay=params['lr_decay'],
        reg_lambda=params['reg_lambda']
    )
    
    # 训练循环
    for epoch in tqdm(range(params['num_epochs']), desc=f"Epochs", leave=False):  # 添加 tqdm
        epoch_loss = 0.0
        for i in range(0, len(X_train), params['batch_size']):
            X_batch = X_train[i:i+params['batch_size']]
            y_batch = y_train[i:i+params['batch_size']]
            model.zero_grad()#最开始的时候清理原来剩余的梯度
            logits = model.forward(X_batch)
            loss, grad = cross_entropy_loss(logits, y_batch, model, trainer.reg_lambda)
            model.backward(grad)
            trainer.step()
            pbar.update(len(X_batch))
        trainer.decay_learning_rate()
                        
    val_acc = evaluate(model, X_val, y_val)
    if val_acc > best_val_acc:
        best_val_acc = val_acc
        best_params = params
        model.save_model(stage=2)
        print(f"\n New best model saved! Validation Acc: {val_acc:.2%} (Params: {params})")
    else:
        # 添加非最佳结果的简洁反馈
        print(f"Current Acc: {val_acc:.2%} (Best: {best_val_acc:.2%})")
        

# New best model saved! Validation Acc: 57.38% 
# (Params: {'reg_lambda': 0.01, 'num_epochs': 100, 'lr_decay': 0.95, 'lr': 0.01, 'hidden_dim2': 512, 'hidden_dim1': 2048, 'batch_size': 64})

# 在测试集上
model.load_model(stage=2)
test_acc = evaluate(model, X_test, y_test)  
print(f"Best Validation Accuracy: {best_val_acc:.2%}")
print(f"Test Accuracy: {test_acc:.2%}")
print(f"Best Parameters: {best_params}")


# 最后输出：
# Current Acc: 44.66% (Best: 58.80%)
# Best Validation Accuracy: 58.80%
# Test Accuracy: 57.33%
# Best Parameters: {'reg_lambda': 0.01, 'num_epochs': 100, 'lr_decay': 0.99, 'lr': 0.01, 'hidden_dim2': 256, 'hidden_dim1': 2048, 'batch_size': 64}
```
